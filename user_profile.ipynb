# Creation a New User Profile

```python

print("\n--- New User Profile ---")

name = input("What is your name? ")
age = int(input("How old are you? "))
married = input("Are you married? (yes/no): ").lower() == "yes"
spouse = input("What is your spouse's name? ") if married else "N/A"
nationality = input("What is your nationality? ")
children = int(input("How many children do you have? "))
current_country = input("Where do you currently live? ")
profession = input("What is your profession? ")
goal = input("What is your main goal in life? ")
studying = input("What are you currently studying? ")

print("\n--- User Profile ---")
print(f"Name: {name}")
print(f"Age: {age}")
print(f"Married: {'Yes' if married else 'No'}")
print(f"Spouse's Name: {spouse}")
print(f"Nationality: {nationality}")
print(f"Number of Children: {children}")
print(f"Current Country: {current_country}")
print(f"Profession: {profession}")
print(f"Main Goal in Life: {goal}")
print(f"Currently Studying: {studying}")    
```
